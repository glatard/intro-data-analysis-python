# Grouping plots

```python
import sklearn.datasets
boston_data = sklearn.datasets.load_boston()
```

```python
import pandas as pd
df = pd.DataFrame(data=boston_data.data, columns=(boston_data.feature_names))
df.head()
```

```python
from matplotlib import pyplot as plt
fig2, axes = plt.subplots(nrows=5, ncols=3, figsize=(15,15))
for i, ax in enumerate(axes.flatten()):
    if i > 12:
        break
    ax.set_ylabel("Target")
    ax.set_xlabel(df.columns[i])
    ax.plot(df[df.columns[i]], boston_data.target, 'o')
```
